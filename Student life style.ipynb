## Drop unused columns and changing values

```python
Student_data['Student_ID'].nunique()
```

```python
Student_data.drop('Student_ID',axis=1,inplace=True)
```

```python
Student_data.drop('Stress_Level',axis=1).corr()
```

```python
Student_data['Stress_Level'].unique()
```

```python
Student_data['Stress_Level'].value_counts()
```

```python
def Stress_level_change(level):
    if level=='Low':
        return 0
    elif level=='Moderate':
        return 1
    elif level=='High':
        return 2
```

```python
Student_data['Stress_Level']=Student_data['Stress_Level'].apply(Stress_level_change)
```

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    #     ..... add as many as you want in your pipeline
    ('std_scaler', StandardScaler()),
])
```

```python
Student_num_tr = my_pipeline.fit_transform(Student_data.drop('GPA',axis=1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
Student_num_tr.shape
```

```python
X_train, X_test, y_train, y_test=train_test_split(Student_num_tr,Student_data['GPA'],random_state=150,test_size=0.3)
```

## Selecting a desired model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predict_result=model.predict(X_test)
```

## Evaluating the model

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, predict_result)
rmse = np.sqrt(mse)
```
